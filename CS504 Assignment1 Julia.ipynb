```julia
# Implementation of the equation
function singlefunction(n)
    a = 1
    for i = 1:n
        a = a*((366-n)/365)
    end
    1-a
end
```

```julia
function repeatedfunction(i)
    A = [ ]
    for n = 1:i
        append!(A,singlefunction(n))
    end
    A
end
```

```julia
#Running Result
repeatedfunction(100)
```

```julia
#Plot
using Plots
plot(repeatedfunction(100),xticks = 0:5:100, yticks = 1:0.1:1)
```
